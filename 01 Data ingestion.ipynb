```python
from src.DataUpdateModule import DataUpdateModule

from datetime import datetime
from dateutil.relativedelta import relativedelta
import holidays, warnings
```

```python
DUM = DataUpdateModule(options_interval_minutes = 5)
nyse_holidays = holidays.NYSE(years = 2023)
```

```python
dfs_stock = dict()
dfs_options = dict()
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    for day in range(365):
        trade_date = datetime(2023, 1, 1) + relativedelta(days = day)
        if (trade_date.weekday() + 1) % 7 in [5] and trade_date not in nyse_holidays:
            print(trade_date)
            a, b = DUM.update_data(trade_date)
            dfs_stock[trade_date] = a
            dfs_options[trade_date] = b
```
